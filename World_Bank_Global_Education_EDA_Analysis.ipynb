##**Data Pre-Processing & Mining for Indicators**
The Data mining is the process of extracting and discovering patterns in large data sets. Data Cleaning is one of the most integral part of analysis, in order to get something meaningful out of a huge dataset we need to figure out how to extract a generalized, smaller and more skewed sub-dataFrame out of it. Like here we have defined two functions, i.e, get_columnwise_data() and get_rowwise_data() to extract out column wise and row wise matrices respectively. Also the function performs some partial cleaning over data such that we only have rows and columns that has some meaningful data to hold. The columns that are all 'NaNs' are dropped such that the data remains cosistent and more skewed. Filling those null columns could effect the analysis, manipulating the graph with values that are not literal and are estimated in one way or the other. Since filling those nulls with a mean or 0, or using forward and backward fill methods, could not make up for the original value, and this is same for all countries so dropping such columns seemed to be the right choice.

```python
# Function to mine a column-wise sub dataset that would have the selected countries as columns
# with meaningful years (those columns/Years that have at least one available value).
# This type of data is suitable for plotting line charts and getting meaningful insights across trends.

def get_columnwise_data(indicator_code, EdStatsData, Countries):
    # Extract data for the given indicator
    df = EdStatsData[EdStatsData["Indicator Code"] == indicator_code]

    # Filter by selected countries
    df = df[df['Country Name'].isin(Countries)]

    # Drop columns with all NaN values
    df.dropna(how='all', axis=1, inplace=True)

    # Reset index
    df.reset_index(drop=True, inplace=True)

    # Extract indicator name safely
    title = df["Indicator Name"].iloc[0] if not df.empty else "Unknown Indicator"

    # Save country names before transposing
    country_names = df["Country Name"].tolist()

    # Drop unnecessary columns before transposing
    cols_to_drop = ["Country Name", "Country Code", "Indicator Name", "Indicator Code"]
    df.drop(columns=[col for col in cols_to_drop if col in df.columns], inplace=True)

    # Transpose and reset index
    df = df.T.reset_index()

    # Define new column names using the extracted country names
    df.columns = ["Year"] + country_names  # âœ… Uses correct country order

    # Drop the first row safely and reset index
    if not df.empty and 0 in df.index:
        df.drop(0, inplace=True)
        df.reset_index(drop=True, inplace=True)

    return df, title


# Function to mine data in a row-wise manner, with years as columns and countries as rows.
# This type of data helps in plotting graphical representations using bars of different heights
# and analyzing trends over time.

def get_rowwise_data(indicator_code, EdStatsData, Countries):
    # Get column-wise data
    df, title = get_columnwise_data(indicator_code, EdStatsData, Countries)

    # Extracting year list for column names
    cols = list(df["Year"])

    # Transpose the dataframe and also reset the indexes of the columns
    df = df.T.reset_index()

    # Setting new column names (Years)
    df.columns = ["Country Name"] + cols

    # Drop the first row safely and reset index
    if not df.empty and 0 in df.index:
        df.drop(0, inplace=True)
        df.reset_index(drop=True, inplace=True)

    return df, title
```

Now, lets just check if the function works as expected. Below is a test run for both functions. We can also use the information retrieved here for future operations on the indicator used.

```python
# Calling get_rowwise_data() on the indicator present at index 1 in the Indicators list.
# This Indicator is described as "Percentage of enrolment in pre-primary education in private institutions (%)".
# This is just to test if the function is returning the required results.
# Call the function with the indicator at index 1
temp, t = get_rowwise_data(Indicators[1], EdStatsData, Countries)

# Print the indicator title
print(t)

# Display the dataframe
temp
```

It can be now seen from the above table that our functions are working exactly as expected. The dataframe returned has all years as columns, that have at least one not-null value. Though the data still has a lot of 'NaN' values, but those we can handle later.

Here we called the get_rowwise_data() function because it implicitly calls the get_columnwise_data() to get a columnwise data matric, which is then converted to rowwise matric by further operation, this way we can test if both the functions are working as desired.
